```python
import os

from openai import OpenAI
```

```python
# Config params
TEMPERATURE = 0.5
MODEL = "gpt-4o-mini"
TOP_P = 0.95

client = OpenAI(
    api_key=os.environ.get(
        "OPENAI_API_KEY", "<your OpenAI API key if not set as env var>"
    )
)
response = client.chat.completions.create(
    model=MODEL,
    temperature=TEMPERATURE,
    top_p=TOP_P,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "What is the capital of the moon?"},
    ],
)
print(response.choices[0].message.content)
```

```python
from datasets import load_dataset

ds = load_dataset("allenai/qasper")
```

```python
import pprint

# Inspect one example format with 2-space indent
pprint.pprint(ds["train"][1], indent=2)
```

```python
# Inspect all extractive questions
count = 0
for qa in ds["train"]:
    for question, answer in zip(
        qa["qas"]["question"], qa["qas"]["answers"], strict=False
    ):
        if count > 10:
            break
        if answer["answer"][0]["extractive_spans"]:
            print(question)
            print(answer["answer"][0]["extractive_spans"])
            print("-" * 100)
            count += 1
```
